# It is a project of Data Analysis with Data Science.

The commands that we used in this project :

* import pandas as pd -- To import Pandas library
* pd.read_csv - To import the CSV file in Jupyter notebook
* head() - It shows the first N rows in the data (by default, N=5)
* shape - It shows the total no. of rows and no. of columns of the dataframe
* df.isnull( ).sum( ) - It detects the missing values from each column of the dataframe.
* fillna() - To fill the null values of a column with some particular value
* value_counts - In a column, it shows all the unique values with their count. It can be applied to a single column only.
* isin() - To show all records including particular elements
* apply() - To apply a function along any axis of DF

# Brand Analysing

```python
fig = plt.figure(figsize = (15, 2))
axes = fig.add_axes([0, 0, 1, 1])
brand_count = df['Make'].value_counts().reset_index()
sns.barplot(x = brand_count['index'], y = brand_count['Make'], )
plt.xticks(rotation = 40)
## count thr top-four's percentage
cnt = 0
for i in range(4):
    cnt += brand_count['Make'][i]
all = brand_count['Make'].sum()
    
axes.axvline(x = 4.5, linestyle = "--", )
lint_text = "Top 4 by count of the total: %s"

percentage = str(round(cnt / all, 4) * 100 ) + "%"
axes.text(2,30.5, (lint_text % percentage))
```

```python
cars = df.groupby(by = 'Origin')['Invoice'].count()
states_top_10 = tuple(cars.sort_values(ascending = False)[:10].index)
df.loc[df['Type'].apply(lambda x:x in states_top_10)]
```

```python
brands = ['Toyota','BMW','Kia','Porsche','Volvo','Jeep','Saab','Lexux','Nissan','Dodge','MINI','Mazda','Jaguar']
fig, ax1 = plt.subplots(figsize=(15,6))

queried = df.loc[df['Origin'].apply(lambda x:x in states_top_10)]\
    .query("Type == 'Sedan' and Make in ['Toyota','BMW','Kia','Porsche','Volvo','Jeep','Saab','Lexux','Nissan','Dodge','MINI','Mazda','Jaguar']")\

data_nation_brand = queried.groupby(by = ['Origin', 'Make'])['Invoice'].count().reset_index()

rates = queried.query("Make == 'Kia'").groupby(by = ['Origin'])['Make'].count() / queried.groupby(by = ['Origin'])['Make'].count()
rates = rates.to_frame().reset_index()
rates['Make'].apply(lambda x: str(round(x * 100 ,4)))

sns.barplot(x = data_nation_brand['Origin'], y = data_nation_brand['Invoice'], hue = data_nation_brand['Make'])
ax2 = ax1.twinx()
sns.lineplot(x = rates['Origin'], y = rates['Make'])
```

# Frequency - Histogram
Histogram refers to the underlying frequency distribution or the probability distribution of a single continuous numerical variable.

# Fuel Consumption:

```python
fig = make_subplots(rows=1, cols=2)

trace_8 = go.Histogram(x=df['MPG_City'].values, name='City MPG', )
trace_9 = go.Histogram(x=df['MPG_Highway'].values, name= 'Highway MPG')

fig.update_layout(title='Fuel Consumption Information', title_x=0.5, yaxis=dict(gridcolor='rgb(255, 255, 255)',gridwidth=3), 
                  paper_bgcolor='rgb(243, 243, 243)', plot_bgcolor='rgb(243, 243, 243)', margin=dict(l=10,r=10,b=50,t=50),
                  height=400, width=900, showlegend=False, hovermode = 'x unified')

# Update xaxis properties

fig.update_xaxes(title_text = 'City MPG', row=1, col=1, nticks=20, ticks="outside",rangemode="tozero",
                showgrid=False,showline=True, linewidth=1, linecolor='gray', tickangle=-90)

fig.update_xaxes(title_text = 'Highway MPG', row=1, col=2, nticks=20, ticks="outside", rangemode="tozero",
                showgrid=False, showline=True, linewidth=1, linecolor='gray', showspikes=True, tickangle=-90)

# Update yaxis properties

fig.update_yaxes(title_text = 'Count', row=1, col=1, nticks=20, ticks="outside")
fig.update_yaxes(row=1, col=2, nticks=15, ticks="outside")

fig.append_trace(trace_8, 1, 1)
fig.append_trace(trace_9, 1, 2)

fig.show()
```

Both City and Highway MPG follows a normal (or Gaussian) distribution.
